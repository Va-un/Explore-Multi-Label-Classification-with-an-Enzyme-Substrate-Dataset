```python
X = df.iloc[:, 0:31]
y = df.iloc[:,31:33]
X_train, X_test,y_train, y_test = train_test_split(X,y , random_state=104,test_size=0.25,shuffle=True)
y1 = y_train['EC1']
y2 = y_train['EC2']
x1 = y_test['EC1']
x2 = y_test['EC2']
Time = []
Time.append("")
```

## Linear Regression

```python
start = time.time()
#EC1
modelL1 = LogisticRegression(C=0.5, penalty='l2', solver='liblinear', max_iter=100)
modelL1.fit(X_train, y1)
predictions = modelL1.predict(X_test)
accuracyL1 = accuracy_score(x1, predictions)
#EC2
modelL2 = LogisticRegression( penalty='l2', solver='liblinear', max_iter=1000)
modelL2.fit(X_train, y2)
predictions = modelL2.predict(X_test)
accuracyL2 = accuracy_score(x2, predictions)
print("Accuracy1:", accuracyL1)
print("Accuracy2:", accuracyL2)
end = time.time()
finalL = (end-start) * 10**3
print("The time of execution of above program is :",finalL, "ms")
Time.append(finalL)
```

## Bagging classifier

```python
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# load the data
seed = 8
kfold = model_selection.KFold(n_splits = 3,
                        random_state = None)


rf_classifier = RandomForestClassifier()
# Set the hyperparameters
rf_classifier.n_estimators = 1000  # Number of decision trees
rf_classifier.criterion = 'gini'  # Splitting criterion
rf_classifier.min_samples_split = 2


# no. of base classifier
num_trees = 10

# bagging classifier
modelB1 = BaggingClassifier(base_estimator = rf_classifier,
                            n_estimators = num_trees,
                            random_state = seed)
modelB1.fit(X_train, y1)

results = model_selection.cross_val_score(modelB1,X_train, y1, cv = kfold)
print("accuracy :")
print(results.mean())
```
